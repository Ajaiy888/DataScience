```python
from transformers import pipeline

chatbot_pipeline = pipeline("text-generation", model="gpt2")
#image_captioning_pipeline = pipeline("image-captioning", model="Salesforce/blip-image-captioning-base")
image_captioning_pipeline = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
!pip install transformers datasets
```

```python
def chatbot_or_image(input_type, text_input, image_input):
    print(f"Input Type: {input_type}, Text Input: {text_input}, Image Input: {image_input}")

    if input_type == "text":
        if text_input:
            response = chatbot_pipeline(text_input, max_length=100, num_return_sequences=1)[0]["generated_text"]
            return f"Chatbot Response: {response}"
        else:
            return "Please provide text input for the chatbot."
    elif input_type == "image":
        if image_input:
            response = image_captioning_pipeline(image_input)
            return f"Image Caption: {response[0]['generated_text']}"
        else:
            return "Please upload an image for captioning."
    else:
        return "Invalid input type selected."
```

```python
def chatbot_or_image(input_type, text_input, image_input):
    if input_type == "text" and not text_input:
        return "Error: No text input provided for the chatbot."
    if input_type == "image" and image_input is None:
        return "Error: No image uploaded for captioning."

```
