```python
# A corpus is a significant collection of texts written in everyday language that computers can read.
import nltk
import nltk.corpus
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
str="Ratan Tata funeral Highlights: ‘Goodbye my lighthouse,’ long-time aide Shantanu Naidu pays final tribute"
str_tokens= word_tokenize(str)
print(str_tokens)
```

```python
type(str_tokens) , len(str_tokens)
```

```python
# frequency
from nltk.probability import FreqDist
wcount = FreqDist()
for word in str_tokens:
  wcount[word]=wcount[word]+1
wcount
```

```python
# top 10 tokens
top_10 = wcount.most_common(10)
top_10
```

```python
# stop_words
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('english')
len(stopwords)
```

```python
print(sorted(stopwords))
```

```python
def extract(sentence):
  words = word_tokenize(sentence)
  word = [word.lower() for word in words if word not in stopwords]
  return word
```

```python
extract(str)
```

```python
# list of all the unique words
def tokenize_unique(sentences):
  words=[]
  for sentence in sentences:
    w = extract(sentence)   # removing the stop words
    words.extend(w)
  words= sorted(list  (set(words)))  # no duplicate
  return words
```

```python
tokenize_unique(str)
```

```python
import numpy as np
def bagofwords(sentence , words):
  words_s = extract(sentence) #remove stop words from sentence

  # count the frequency
  bag = np.zeros(len(words))
  for sw in words_s:
    for i , word in enumerate(words):
      if word == sw:
        bag[i]+=1
  return np.array(bag)
```

```python
sentences = ["learning python " ,
             " ria learning python for data analytics" ,
             " data analytics is filed of data science "]
bag_of_words =tokenize_unique(sentences)
print(bag_of_words)
```

```python
bagofwords("ria learning java learning python " , bag_of_words)
```
